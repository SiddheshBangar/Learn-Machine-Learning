### ðŸ“Œ One Hot Encoding with top categories

```python
counts = df['brand'].value_counts()
threshold = 100
```

```python
replce = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(replce, 'uncommon'))
```
