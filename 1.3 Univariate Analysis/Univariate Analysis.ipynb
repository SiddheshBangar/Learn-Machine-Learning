# Univariate Analysis

The term univariate analysis refers to the analysis of one variable. You can remember this because the prefix **‚Äúuni‚Äù** means **‚Äúone.‚Äù**

Univariate analysis is a statistical analysis technique used to analyze a single variable at a time. In other words, it involves analyzing one variable in isolation, without considering the relationship between that variable and other variables.

In univariate analysis, the focus is on describing and understanding the characteristics of a single variable, such as its distribution, central tendency, variability, and outliers. Common techniques used in univariate analysis include measures of central tendency (such as mean, median, and mode), measures of variability (such as range, standard deviation, and variance), and graphical methods (such as histograms, box plots, and density plots).

Univariate analysis is often used as a first step in statistical analysis, as it provides a basic understanding of the data before more complex analyses are conducted. It can also be used to identify potential outliers or data errors, and to check whether the data meet certain assumptions of statistical tests.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('train.csv')
```

## Categorical Data

Categorical data is data that is collected in groups or topics; the number of events in each group is counted numerically.

### üìå Countplot

The countplot is used to represent the occurrence(counts) of the observation present in the categorical variable.

It uses the concept of a bar chart for the visual depiction.

```python
sns.countplot(data=df, x='Pclass', hue='Survived', palette='bright')
```

### üìå Pie Chart

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. 

```python
df['Pclass'].value_counts().plot(kind='pie')
```

## Numerical Data

Numerical data is a data type expressed in numbers, rather than natural language description. Sometimes called quantitative data, numerical data is always collected in number form. Numerical data differentiates itself from other number form data types with its ability to carry out arithmetic operations with these numbers.

### üìå Histogram

A histogram is an approximate representation of the distribution of numerical data. 

```python
plt.hist(df['Age'], color='green')
```

### üìå Distplot

A Distplot or distribution plot, depicts the variation in the data distribution.

```python
sns.distplot(df['Age'], color='red')
```

### üìå Boxplot

When we display the data distribution in a standardized way using 5 summary ‚Äì minimum, Q1 (First Quartile), median, Q3(third Quartile), and maximum, it is called a Box plot. It is also termed as box and whisker plot. 

```python
sns.boxplot(x=df['Pclass'], y=df['Age'], palette='bright')
```
