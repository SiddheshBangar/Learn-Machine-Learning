# âš¡ Handling Mixed Variables

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
df.sample(5)
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers Travelling with')
```

```python
# Extract Numerical part
df['numerical_number'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
df['number_categorical'] = np.where(df['numerical_number'].isnull(), df['number'], np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['Cabin_num'] = df['Cabin'].str.extract('(\+d)') # Captures numerical part
df['Cabin_cat'] = df['Cabin'].str[0] # Captures categorical part

df.sample(5)
```

```python
df['Cabin_cat'].value_counts()
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
